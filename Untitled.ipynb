```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {'Employee id': [10, 20, 15, 25, 30, 34, 54, 54],
        'Gender': ['M', 'F', 'F', 'M', 'F','M','M','F'],
        'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice','Bad','Nice','Bad'],
        }

df = pd.DataFrame(data)
print(f"Employee data : \n{df}")
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Gender', 'Remarks']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Gender', 'Remarks']))
df_encoded = pd.concat([df.drop(columns=['Gender', 'Remarks']), encoded_df], axis=1)

print(f"Encoded Employee data : \n{df_encoded}")
```
