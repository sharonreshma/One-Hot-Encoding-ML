```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {
    'Employee id': [10, 20, 15, 25, 30, 34, 54, 54],
    'Gender': ['M', 'F', 'F', 'M', 'F', 'M', 'M', 'F'],
    'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice', 'Bad', 'Nice', 'Bad'],
}

df = pd.DataFrame(data)
print(f"Employee data : \n{df}")

encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['Gender', 'Remarks']])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['Gender', 'Remarks']))

df_encoded = pd.concat([df[['Employee id']], encoded_df], axis=1)

print(f"\nEncoded Employee data : \n{df_encoded}")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv(r"C:\One-Hot-Encoding-ML\Emp_data.csv")
print("Original Employee data:\n", df)

categorical_columns = ['GENDER', 'REMARK']
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[categorical_columns])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(categorical_columns))
df_encoded = pd.concat([df, encoded_df], axis=1)

print("\nEncoded Employee data:\n", df_encoded)
```
